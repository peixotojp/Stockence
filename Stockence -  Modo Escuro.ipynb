```python
import pandas as pd
import pandas_datareader as pdr
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rcParams
```

```python
ação = 'ITUB4.SA'
açãoum = 'BBDC4.SA'
açãodois = 'SANB4.SA'
itub4 = pdr.get_data_yahoo(ação, start = '2020-08-19')['Close']
bbdc4 = pdr.get_data_yahoo(açãoum, start = '2020-08-19')['Close']
sanb4 = pdr.get_data_yahoo(açãodois, start = '2020-08-19')['Close']
```

```python
plt.figure(figsize=(20, 10))
plt.plot(itub4, color='#ff9100')
plt.plot(bbdc4, color= '#ff0000')
plt.plot(sanb4, color='#b90000')
plt.yticks(np.arange(min(sanb4)-1, max(itub4)+1, 0.5))
plt.grid(True)
plt.legend(['ITUB4','BBDC4','SANB4'], loc='center left', prop={'size': 15})
plt.title('ITUB4 vs BBDC4 vs SANB4 - 19/08/2020 - 21/09/2020')
plt.savefig('Data/gráficoumescuro.png')
rcParams['font.family'] = 'Arial'
rcParams['legend.facecolor'] = 'w'
rcParams['xtick.labelsize'] = 15
rcParams['ytick.labelsize'] = 15
rcParams['grid.color'] = 'w'
rcParams['axes.facecolor'] = '#303030'
```
