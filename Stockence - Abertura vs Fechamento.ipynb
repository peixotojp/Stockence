```python
import pandas as pd
import pandas_datareader as pdr
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
ação = 'TRIS3.SA'
itub4 = pdr.get_data_yahoo(ação, start = '2020-08-19')['Open']
itub = pdr.get_data_yahoo(ação, start = '2020-08-19')['Close']
```

```python
rcParams['font.family'] = 'Arial'
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
plt.figure(figsize=(20, 10))
plt.grid()
plt.plot(itub4)
plt.plot(itub)
plt.legend(['Abertura', 'Fechamento'], loc='best', prop={'size': 15})
plt.title('ITUB4 - Abertura vs Fechamento')
```
