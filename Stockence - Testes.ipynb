```python
import pandas as pd
import numpy as np
import pandas_datareader as pdr
import matplotlib.pyplot as plt
import matplotlib.ticker as plticker
from matplotlib import rcParams
```

```python
ação = 'ITUB4.SA'
açãoum = 'BBDC4.SA'
açãodois = 'SANB4.SA'
itub4 = pdr.get_data_yahoo(ação, start = '2020-08-19')['Close']
bbdc4 = pdr.get_data_yahoo(açãoum, start = '2020-08-19')['Close']
sanb4 = pdr.get_data_yahoo(açãodois, start = '2020-08-19')['Close']
```

```python
rcParams['font.family'] = 'Arial'
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
plt.figure(figsize=(20, 10))
plt.yticks(np.arange(min(sanb4), max(itub4)+1, 0.5))
plt.plot(itub4, color='#ff9100', lw=2.5)
plt.plot(bbdc4, color= '#ff0000', lw=2.5)
plt.plot(sanb4, color='#b90000', lw=2.5)
plt.legend(['ITUB4','BBDC4','SANB4'], loc='best', prop={'size': 15})
plt.grid(True)
plt.title('ITUB4 vs BBDC4 vs SANB4 - 19/08/2020 - 19/10/2020')
plt.savefig('Data/graficoumatualizado.png')
```
